```python
def calculate_total_distance(km_start, km_end):
    # Convert strings to numerical values
    start = [float(s.replace('+', '.')) for s in km_start]
    end = [float(e.replace('+', '.')) for e in km_end]
    # Calculate the differences and sum them
    return sum(e - s for s, e in zip(start, end))

# Apply the function to calculate total distance
grouped_data['total_distance'] = grouped_data.apply(lambda row: calculate_total_distance(row['km_start'], row['km_end']), axis=1)
total_dist_truck = grouped_data.drop(columns=['km_start', 'km_end',id])

# Display the resulting DataFrame
total_dist_truck
```

```python
from leuvenmapmatching.matcher.distance import DistanceMatcher
from leuvenmapmatching.map.inmem import InMemMap

def match_path_to_graph(path, graph):
    map_con = InMemMap("myosm", use_latlon=True, use_rtree=True, index_edges=True)

    nodes_id = list(graph.nodes)
    for node in nodes_id:
        lat = graph.nodes[node]['y']
        lon = graph.nodes[node]['x']
        map_con.add_node(node, (lat, lon))

    edges_id = list(graph.edges)
    for edge in edges_id:
        node_a, node_b = edge[0], edge[1]
        map_con.add_edge(node_a, node_b)
        map_con.add_edge(node_b, node_a)  # Add reverse edge

    map_con.purge()

    matcher = DistanceMatcher(map_con, max_dist=30, obs_noise=20, min_prob_norm=0.5,
                              non_emitting_states=True, only_edges=True)
    states, _ = matcher.match(path)
    nodes = matcher.path_pred_onlynodes

    return states, nodes

states, nodes = match_path_to_graph(traj, graph)
```

##### Algorithm 6.1

```python
def obtain_route_edges(route_waypoints,places):
    """
    To obtain a set of route edges.

    :param route_waypoints: List of waypoints, where each waypoint is represented as (Lat, Lng)
    :return: Set of route edges
    """
    M = len(route_waypoints)
    route_edges = []
    cumulative_distance = 0  # Initialize cumulative distance

    for i in range(M - 1):
        # Extract latitude and longitude for points i and i+1
        Lat_i, Lng_i = route_waypoints[i]
        Lat_next, Lng_next = route_waypoints[i + 1]
        road_intersect_1 = places[i]
        road_intersect_2 = places[i+1]

        # Calculate slope with zero-division handling
        if Lng_next == Lng_i:
            m_i = float('inf')  # Use infinity to represent a vertical slope
        else:
            m_i = (Lat_next - Lat_i) / (Lng_next - Lng_i)

        # Calculate the distance between the two points
        l_i = haversine_distance(route_waypoints[i], route_waypoints[i + 1])

        # Update cumulative distance
        cumulative_distance += l_i

        # Create the route edge with required attributes
        edge = (route_waypoints[i], route_waypoints[i + 1], m_i, l_i, cumulative_distance, road_intersect_1, road_intersect_2)
        route_edges.append(edge)

    # Convert the list of edges to the required output format
    return route_edges




def haversine_distance(point1, point2):
    """
    Calculate the great-circle distance between two points on the Earth using the Haversine formula.

    :param point1: Tuple (Lat1, Lng1) in degrees
    :param point2: Tuple (Lat2, Lng2) in degrees
    :return: Distance in kilometers
    """
    R = 6371  # Earth's radius in kilometers

    # Convert latitude and longitude from degrees to radians
    Lat1, Lng1 = map(radians, point1)
    Lat2, Lng2 = map(radians, point2)

    # Haversine formula
    dLat = Lat2 - Lat1
    dLng = Lng2 - Lng1
    a = sin(dLat / 2)**2 + cos(Lat1) * cos(Lat2) * sin(dLng / 2)**2
    c = 2 * asin(sqrt(a))
    distance = R * c
    # Distance in kilometers
    return distance

```
